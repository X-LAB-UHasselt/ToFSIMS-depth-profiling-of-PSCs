Since Google Scholar search results are dynamic, we updated the values to the results obtained on 19/02/2026.

```python
# Searched using Google Scholar: https://scholar.google.com/
# We always inspect "About X results" and analyze X
# How to use Google Scholar searching: https://library.acg.edu/how-to-guides/google-scholar/advanced-searching

years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

# Google Scholar "perovskite solar cell" (advanced search: with all of the words, anywhere in the article, dated between year-year)
number_of_publications_perovskite = [4130, 7690, 11300, 13900, 17400, 20200, 23100, 27400, 31300, 33700, 38200, 34900]

# Google Scholar "ToF-SIMS" (advanced search: with all of the words, anywhere in the article, dated between year-year)
number_of_publications_ToFSIMS = [2580, 2730, 2860, 2920, 3090, 3270, 3520, 3840, 4190, 4460, 5050, 7180]

# Google Scholar "perovskite solar cell AND ToF-SIMS" (advanced search: with all of the words, anywhere in the article, dated between year-year)
number_of_publications_perovskite_and_ToFSIMS = [40, 70, 113, 188, 227, 319, 424, 512, 626, 719, 914, 1270]

# Relative amount of ToF-SIMS in perovskite and vice versa
relative_perovskite_and_ToFSIMS_vs_perovskite = [100*i/j for i, j in zip(number_of_publications_perovskite_and_ToFSIMS, number_of_publications_perovskite)]
relative_perovskite_and_ToFSIMS_vs_ToFSIMS = [100*i/j for i, j in zip(number_of_publications_perovskite_and_ToFSIMS, number_of_publications_ToFSIMS)]

# Relative ToF-SIMS @ PVK in 2024 vs 2015
print("{:.2f}".format(relative_perovskite_and_ToFSIMS_vs_perovskite[-1]/relative_perovskite_and_ToFSIMS_vs_perovskite[1]))

fig, ax = plt.subplots(1, 1, figsize=(4, 3), constrained_layout=True)
ax.plot(years, relative_perovskite_and_ToFSIMS_vs_perovskite, "o-", c="k")
ax.set_xlabel("Year")
ax.set_ylabel("Relative ToF-SIMS in perovskite (%)")
ax.grid()
print(max(relative_perovskite_and_ToFSIMS_vs_perovskite))
# location = './../output/misc/publication_statistics/'
# filename = 'statistics_GS_relative-ToFSIMS-in-perovskite'
# extension = '.png'
# fig.savefig(fname=location+filename+extension, dpi=1200, format='png', transparent=None, bbox_inches='tight')

fig, ax = plt.subplots(1, 1, figsize=(4, 3), constrained_layout=True)
ax.plot(years, relative_perovskite_and_ToFSIMS_vs_ToFSIMS, "o-", c="k")
ax.set_xlabel("Year")
ax.set_ylabel("Relative perovskite in ToF-SIMS (%)")
ax.grid()
print(max(relative_perovskite_and_ToFSIMS_vs_ToFSIMS))
# location = './../output/misc/publication_statistics/'
# filename = 'statistics_GS_relative-perovskite-in-ToFSIMS'
# extension = '.png'
# fig.savefig(fname=location+filename+extension, dpi=1200, format='png', transparent=None, bbox_inches='tight')
```
